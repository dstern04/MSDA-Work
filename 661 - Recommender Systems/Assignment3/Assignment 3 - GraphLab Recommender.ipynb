Before loading the data, I formatted the ratings a bit. The ratings took on values from (0,5] at 0.5 intervals. The graphlab.recommender.create function requires that ratings be integers or strings that it can factorize, so I multiplied all of the ratings values by 10.

Now we can create recommenders that uses item-item similarities based on users in common. We train this next model using only the user and item data. 

```python
model2 = graphlab.recommender.item_similarity_recommender.create(data, user_id='"userId"', item_id='"title"')
```

```python
recs2 = model2.recommend()
recs2[recs2['"userId"'] == 168]
```

We can enhance the model above by incorporating the ratings data to enhance the measure of item-item similarities. The similarity_type parameter defaults to jaccard, but we can also choose from cosine and pearson measures. Here we choose the pearson correlation coefficient. 

```python
model3 = graphlab.recommender.item_similarity_recommender.create(data, user_id='"userId"', item_id='"title"',
                                                                 target='"rating"', similarity_type='pearson')
```

```python
recs3 = model3.recommend()
recs3[recs3['"userId"'] == 168]
```
